```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import os
import time
import datetime

import numpy as np
import random
from tqdm import tqdm
import torch
from torch.utils.data import Dataset, DataLoader, random_split, RandomSampler

from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import AdamW, get_linear_schedule_with_warmup

# Cargar el modelo y el tokenizador
model_name = 'gpt2'  # Puedes cambiar esto a 'gpt2-medium', 'gpt2-large', etc.
tokenizer = AutoTokenizer.from_pretrained("flax-community/gpt-2-spanish", bos_token='<|startoftext|>', eos_token='<|endoftext|>', pad_token='<|pad|>')
model = AutoModelForCausalLM.from_pretrained("flax-community/gpt-2-spanish")

# Función para generar texto
def generate_text(prompt, max_length=50):
    # Tokenizar la entrada
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generar texto
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1)

    # Decodificar y retornar el texto generado
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Usar la función
prompt = "En Colombia el dólar cerró abajo de $4.200 tras los datos de inflación subyacente en Estados Unidos. Esto se puede explicar por "
generated_text = generate_text(prompt, max_length=200)
print(generated_text)
```
