```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import AutoModelForCausalLM, AutoTokenizer

import tensorflow as tf

tokenizer = AutoTokenizer.from_pretrained("datificate/gpt2-small-spanish")
model = AutoModelForCausalLM.from_pretrained("datificate/gpt2-small-spanish")
```

```python
sequence = "En Colombia el dólar cerró abajo de $4.200 tras los datos de inflación subyacente en Estados Unidos. Esto se puede explicar por "
```

```python
inputs = tokenizer.encode(sequence, return_tensors='pt')
```

```python
# we pass a maximum output length of 200 tokens
outputs = model.generate(inputs, max_length=100, do_sample=True)
```

```python
text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
outputs = model.generate(inputs, max_length=120, do_sample=True)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
outputs = model.generate(inputs, max_length=120, do_sample=True, temperature=0.7)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
outputs = model.generate(inputs, max_length=120, do_sample=True, top_k=50)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```
